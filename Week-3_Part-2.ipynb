### Reading the data posted on the website

```python
import pandas as pd # library for data analsysis
import numpy as np # library to handle data in a vectorized manner

link = "http://cocl.us/Geospatial_data"
df1 = pd.read_csv(link)

df1.head()
```

```python
df1.shape
```

### Column names with "Postal Code" changed to "Postcode" 

```python
df1.columns = ['Postcode','Latitude','Longitude']

cols = df1.columns.tolist()
cols
```

### Reading the file which is posted in Github Repository

```python
link = "https://github.com/Sahar-TJ/Coursera-Capstone/blob/master/Geospatial_Coordinates.csv"
df = pd.read_csv(link,index_col=0)
df.head()
```

```python
df.shape
```

### Merging the Two Tables

```python
df_new = pd.merge(df, df1, on='Postcode')
df_new
```
